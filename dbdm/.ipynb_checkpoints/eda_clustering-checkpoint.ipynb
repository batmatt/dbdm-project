```python
import pandas as pd

with open('../data/kym.json', 'r') as file:
    data = pd.read_json(file)

data.head()
```

```python
# Status distribution
status_dict = {}
c = 0
for d in data['details']:
    if 'status' in d.keys():
        status = d['status']
        if status in status_dict.keys():
            status_dict[status] += 1
        else:
            status_dict[status] = 1
            
print("Status distribution:", status_dict)
```

```python
# Count occurrences of each unique category
category_counts = data['category'].value_counts()

import matplotlib.pyplot as plt

# Plotting a bar plot for category counts
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.show()

```

```python
from collections import Counter

year_column = []
missing_year = []
for title, d in zip(data['title'], data['details']):
    if 'year' in d.keys() and d['year'] != None:
        year_column.append(d['year'])
    else:
        year_column.append("")

data['year'] = year_column

year_freq = Counter(data['year'])
text = "YEAR is missing for {} memes out of {}.".format(year_freq[''], len(data))
print(text)

# Create bar plot
plot_data = dict(sorted(year_freq.items(), reverse=True))
plot_data = {str(key): val for key, val in plot_data.items() if int(val) >= 30}  # keep only items with at least 30 occurrences
plot_data = dict(sorted(plot_data.items(), reverse=True))

if "" in plot_data.keys():
    plot_data.pop("")

# Adjust figure size and font sizes
plt.figure(figsize=(12, 6))
plt.bar(plot_data.keys(), plot_data.values(), color="green")
plt.ylabel('Total number', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.title("Year distribution", fontsize=20)
plt.xticks(rotation=45, ha="right", fontsize=10)  # Adjust rotation and fontsize
plt.yticks(fontsize=10)

# Set ylim based on data range
plt.ylim([0, max(plot_data.values()) + 10])

# Tight layout for better spacing
plt.tight_layout()

plt.show()
```

```python
small_fontsize=10
medium_fontsize=15
large_fontsize=20

all_types = []

# Iterate over the 'details' column and collect types
for d in data['details']:
    types = d.get('type', [])
    all_types.extend([item.replace('https://knowyourmeme.com/types/', '') if item else "" for item in types])

# Count occurrences of each type
type_counts = Counter(all_types)

# Filter types with occurrences larger than 30
filtered_type_counts = {key: value for key, value in type_counts.items() if value > 50}

# Create a list of colors, painting red for specified types
selected_types = {'exploitable', 'reaction', 'snowclone', 'catchphrase'}
colors = ['green' if key not in selected_types else 'red' for key in filtered_type_counts.keys()]

# Create a histogram
plt.figure(figsize=(15, 8))
bars = plt.bar(filtered_type_counts.keys(), filtered_type_counts.values(), color=colors)
plt.xlabel('Type', fontsize=medium_fontsize)
plt.ylabel('Occurrences', fontsize=medium_fontsize)
plt.title('Occurrences of Each Type (Count > 50)', fontsize=large_fontsize)
plt.xticks(rotation=90, fontsize=small_fontsize)
plt.yticks(fontsize=small_fontsize)

# Add legend for color-coding
legend_labels = ['Other Types', 'Exploitable, Reaction, Snowclone, Catchphrase']
plt.legend(bars, legend_labels, loc='upper right', fontsize=small_fontsize)

plt.tight_layout()
plt.show()
```

```python
# alter details table to drop the 'https://knowyourmeme.com/types/' preamble for types
data['details'] = data['details'].apply(lambda details: {'type': [t.replace('https://knowyourmeme.com/types/', '') for t in details.get('type', [])]} if 'type' in details else details)

# Create a mask to filter entries
mask = data['details'].apply(lambda details: any(type_ in selected_types for type_ in details.get('type', [])))

# Filter the DataFrame based on the mask
selected_entries = data[mask]

# Display the selected entries
print(selected_entries.head())

# Count the number of entries matching the selected types
count_selected_entries = len(selected_entries)
print("Number of entries with selected types:", count_selected_entries)
```

```python
# Extract tags from the 'tags' column
all_tags = [tag for taglist in selected_entries["tags"] for tag in taglist]

# Count missing tags
missing_tags = sum(1 for taglist in selected_entries["tags"] if len(taglist) == 0)

# Count tag occurrences
tags_freq = Counter(all_tags)

# Display information about missing tags and unique tags
text = "TAGS is missing for {} memes out of {}. There are {} different tags. Showing only tags with >=3 occurrences. Remember: some memes have multiple tags."\
      .format(missing_tags, len(selected_entries), len(tags_freq))

# Filter tags with occurrences >= 50
filtered_tags = {key: value for key, value in tags_freq.items() if value >= 50}

# Create bar plot
plt.figure(figsize=(15, 8))
plt.bar(filtered_tags.keys(), filtered_tags.values(), color='green')
plt.ylabel('Total number', fontsize=medium_fontsize)
plt.xlabel('Tag', fontsize=medium_fontsize)
plt.title("Tag distribution", fontsize=large_fontsize)
plt.xticks(rotation=45, fontsize=small_fontsize)
plt.yticks(fontsize=small_fontsize)
plt.xlim([-1, len(filtered_tags.keys())])

# Show the plot
plt.show()
```

```python
# Get all descriptions
text_values = selected_entries["content"].apply(lambda x: str(x.get("about", {}).get("text", "")) if isinstance(x, dict) and x.get("about") and x["about"].get("text") else "")

# Remove entries with no text value and duplicate text values
selected_entries = selected_entries[text_values != ""].drop_duplicates(subset='content', keep=False)

# Display the resulting DataFrame without entries with no text value and duplicate text values
print(f"Number of entries after removing duplicates and those with no text value: {selected_entries.shape[0]}")
text_values = selected_entries["content"].apply(lambda x: str(x.get("about", {}).get("text", "")) if isinstance(x, dict) and x.get("about") and x["about"].get("text") else "")

print(text_values)
```

```python
from wordcloud import WordCloud

# Transform list of descriptions of selected entires to one string
all_text = text_values.str.cat(sep=' ')

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for "Description" section of memes of type {Exploitable, Reaction, Catchphrase, Snowclone}', fontsize=15)
plt.show()
```

```python
### To cluster memes based on descriptions, you can use natural language processing (NLP) techniques. One common approach is to use vectorization methods such as TF-IDF (Term Frequency-Inverse Document Frequency) to convert text data into numerical vectors and then apply a clustering algorithm.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

descriptions = text_values

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(descriptions)

# K-Means Clustering
num_clusters = 3  # You can choose the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Add cluster labels to the selected_entries DataFrame
selected_entries["cluster_label"] = kmeans.labels_

# Display the resulting DataFrame with cluster labels
print(selected_entries[["content", "cluster_label"]])
```

```python
numClusters = range(1, 25)
SSE = []
for k in numClusters:
    k_means = KMeans(n_clusters=k, n_init='auto')
    k_means.fit(X)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Transform the vectorized data using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Add PCA components to the selected_entries DataFrame
x0 = X_pca[:, 0]
x1 = X_pca[:, 1]
selected_entries["PCA1"] = x0
selected_entries["PCA2"] = x1

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x="PCA1", y="PCA2", hue="cluster_label", data=selected_entries, palette="viridis", alpha=0.7)
plt.title("Meme Clusters based on Descriptions")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster Label", loc="upper right")
plt.show()
```

```python
with open('../data/kym.json', 'r') as file:
    data = pd.read_json(file)

data['cleaned_type'] = ""

# Iterate through rows
for index, row in data.iterrows():
    types = row['details'].get('type', [])
    cleaned_types = [t.replace('https://knowyourmeme.com/types/', '') for t in types]
    data.at[index, 'cleaned_type'] = cleaned_types

# Count instances with empty 'cleaned_type'
empty_type_count = sum(len(row['cleaned_type']) == 0 for index, row in data.iterrows())
print(f"Number of entries with 'cleaned_type': {len(data) - empty_type_count}")
print(f"Number of entries with empty 'cleaned_type': {empty_type_count}")

# Remove entries with empty 'cleaned_type'
data = data[data['cleaned_type'].apply(len) > 0]


data['cleaned_description'] = ""

# Initialize a variable to store the count of rows with float values
float_rows_count = 0

# Iterate through rows
for index, row in data.iterrows():
    content_value = row['content']

    if isinstance(content_value, dict):
        about_text = content_value.get('about', {}).get('text', [])
        cleaned_description = ' '.join(about_text)  # Combine text into a single string
        data.at[index, 'cleaned_description'] = cleaned_description
    else:
        float_rows_count += 1
        data.at[index, 'cleaned_description'] = ''

# Drop rows where 'content' is of type float
data = data.dropna(subset=['content'])

print(f"Number of rows dropped with float 'content': {float_rows_count}")
data.head()

# Flatten the 'cleaned_type' lists
all_types = [item for sublist in data['cleaned_type'] for item in sublist]

# Create a DataFrame for counting occurrences
types_df = pd.DataFrame({'type': all_types})

# Plot a histogram
plt.figure(figsize=(10, 6))
types_df['type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Histogram of Occurrences by Cleaned Type')
plt.xlabel('Cleaned Type')
plt.ylabel('Occurrences')
plt.show()
```
